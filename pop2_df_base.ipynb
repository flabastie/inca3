### Table DESCRIPTION_INDIV : Sélection variables & sélection population adulte

```python
# Table DESCRIPTION_INDIV
path_description_indiv = "./datas/description-indiv.csv"
df_description_indiv = pd.read_csv(path_description_indiv, sep=';', error_bad_lines=False, index_col=False)

# Sélection populations (conditions)
pop2 = df_description_indiv['pop2'] == 1
#  select pop2
df_description_indiv_pop_2 = df_description_indiv[pop2]

# Sélection variables
description_indiv_columns = ['NOIND', 'pop2', 'pop3', 'agglo_5cl', 'sex_PS', 'tage_PS', 'diplome_interv', 'revenu', 'imc',  'fume']
df_description_indiv_select = df_description_indiv_pop_2[description_indiv_columns]

# Sélection classes adultes
print('df_description_indiv.shape :', df_description_indiv.shape)
is_adult = df_description_indiv_select['tage_PS'].isin(['7','8','9'])
df_description_indiv_select = df_description_indiv_select[is_adult]
print('df_description_indiv_select.shape :', df_description_indiv_select.shape)

# Traitement valeurs manquantes
df_description_indiv_select['imc'].fillna(df_description_indiv_select['imc'].mean(), inplace=True)
# Traitement valeurs manquantes: sex_PS == NaN => 1 (homme)
df_description_indiv_select.loc[df_description_indiv_select['sex_PS'].isna(), 'sex_PS'] = 1
# Traitement valeurs manquantes: tage_PS == NaN => 8 (45-64ans)
df_description_indiv_select.loc[df_description_indiv_select['tage_PS'].isna(), 'tage_PS'] = 8
# Traitement valeurs manquantes: agglo_5cl == NaN => 5 (Agglo Paris)
df_description_indiv_select.loc[df_description_indiv_select['agglo_5cl'].isna(), 'agglo_5cl'] = 5

# traitement diplome_interv - Valeurs manquantes et Refus / Ne sait pas
df_description_indiv_select.loc[df_description_indiv_select['diplome_interv'].isna(), 'diplome_interv'] = 0
df_description_indiv_select['diplome_interv'] = df_description_indiv_select['diplome_interv'] + 10
df_description_indiv_select.loc[df_description_indiv_select['diplome_interv'].isin(['10','11','15','23','24']), 'diplome_interv'] = 1
df_description_indiv_select.loc[df_description_indiv_select['diplome_interv'].isin(['12','13','14','16']), 'diplome_interv'] = 2
df_description_indiv_select.loc[df_description_indiv_select['diplome_interv'] == 17, 'diplome_interv'] = 3
df_description_indiv_select.loc[df_description_indiv_select['diplome_interv'] == 18, 'diplome_interv'] = 4
df_description_indiv_select.loc[df_description_indiv_select['diplome_interv'] == 19, 'diplome_interv'] = 5
df_description_indiv_select.loc[df_description_indiv_select['diplome_interv'] == 20, 'diplome_interv'] = 6
df_description_indiv_select.loc[df_description_indiv_select['diplome_interv'] == 21, 'diplome_interv'] = 7
df_description_indiv_select.loc[df_description_indiv_select['diplome_interv'] == 22, 'diplome_interv'] = 8

# traitement revenu - Valeurs manquantes et Refus / Ne sait pas
df_description_indiv_select.loc[df_description_indiv_select['revenu'].isna(), 'revenu'] = 1
df_description_indiv_select.loc[df_description_indiv_select['revenu'] == 14, 'revenu'] = 1
df_description_indiv_select.loc[df_description_indiv_select['revenu'] == 15, 'revenu'] = 1

# traitement fume
df_description_indiv_select.loc[df_description_indiv_select['fume'].isna(), 'fume'] = 0
df_description_indiv_select['fume'] = df_description_indiv_select['fume'] + 10
df_description_indiv_select.loc[df_description_indiv_select['fume'] == 14, 'fume'] = 0
df_description_indiv_select.loc[df_description_indiv_select['fume'] == 13, 'fume'] = 1
df_description_indiv_select.loc[df_description_indiv_select['fume'] == 12, 'fume'] = 2
df_description_indiv_select.loc[df_description_indiv_select['fume'] == 11, 'fume'] = 3
df_description_indiv_select.loc[df_description_indiv_select['fume'] == 10, 'fume'] = 0

print("Missing values = ", df_description_indiv_select.isna().sum().sum())
# print(df_description_indiv_select.describe())
df_description_indiv_select.head()
```

## Table HABITUDES_INDIV : Sélection variables



```python
# Table HABITUDES_INDIV
path_habitudes_indiv = "./datas/habitudes-indiv.csv"
df_habitudes_indiv = pd.read_csv(path_habitudes_indiv, sep=';', error_bad_lines=False, index_col=False)

# formatage_bio - df_habitudes_indiv
def formatage_bio(col_name):
  df_habitudes_indiv.loc[df_habitudes_indiv[col_name].isna(), col_name] = 0
  df_habitudes_indiv[col_name] = df_habitudes_indiv[col_name] + 10
  df_habitudes_indiv.loc[df_habitudes_indiv[col_name] == 16, col_name] = 0
  df_habitudes_indiv.loc[df_habitudes_indiv[col_name] == 15, col_name] = 0
  df_habitudes_indiv.loc[df_habitudes_indiv[col_name] == 14, col_name] = 0
  df_habitudes_indiv.loc[df_habitudes_indiv[col_name] == 13, col_name] = 1
  df_habitudes_indiv.loc[df_habitudes_indiv[col_name] == 12, col_name] = 2
  df_habitudes_indiv.loc[df_habitudes_indiv[col_name] == 11, col_name] = 3
  df_habitudes_indiv.loc[df_habitudes_indiv[col_name] == 10, col_name] = 0

# traitement restaurationrapide_freq
df_habitudes_indiv.loc[df_habitudes_indiv['restaurationrapide_freq'].isna(), 'restaurationrapide_freq'] = 0
df_habitudes_indiv['restaurationrapide_freq'] = df_habitudes_indiv['restaurationrapide_freq'] + 10
df_habitudes_indiv.loc[df_habitudes_indiv['restaurationrapide_freq'] == 18, 'restaurationrapide_freq'] = 0
df_habitudes_indiv.loc[df_habitudes_indiv['restaurationrapide_freq'] == 17, 'restaurationrapide_freq'] = 0
df_habitudes_indiv.loc[df_habitudes_indiv['restaurationrapide_freq'] == 16, 'restaurationrapide_freq'] = 1
df_habitudes_indiv.loc[df_habitudes_indiv['restaurationrapide_freq'] == 15, 'restaurationrapide_freq'] = 2
df_habitudes_indiv.loc[df_habitudes_indiv['restaurationrapide_freq'] == 14, 'restaurationrapide_freq'] = 3
df_habitudes_indiv.loc[df_habitudes_indiv['restaurationrapide_freq'] == 13, 'restaurationrapide_freq'] = 4
df_habitudes_indiv.loc[df_habitudes_indiv['restaurationrapide_freq'] == 12, 'restaurationrapide_freq'] = 5
df_habitudes_indiv.loc[df_habitudes_indiv['restaurationrapide_freq'] == 11, 'restaurationrapide_freq'] = 6
df_habitudes_indiv.loc[df_habitudes_indiv['restaurationrapide_freq'] == 10, 'restaurationrapide_freq'] = 0

# traitement collation_freq
df_habitudes_indiv.loc[df_habitudes_indiv['collation_freq'].isna(), 'collation_freq'] = 0
df_habitudes_indiv['collation_freq'] = df_habitudes_indiv['collation_freq'] + 10
df_habitudes_indiv.loc[df_habitudes_indiv['collation_freq'] == 17, 'collation_freq'] = 0
df_habitudes_indiv.loc[df_habitudes_indiv['collation_freq'] == 16, 'collation_freq'] = 0
df_habitudes_indiv.loc[df_habitudes_indiv['collation_freq'] == 15, 'collation_freq'] = 1
df_habitudes_indiv.loc[df_habitudes_indiv['collation_freq'] == 14, 'collation_freq'] = 2
df_habitudes_indiv.loc[df_habitudes_indiv['collation_freq'] == 13, 'collation_freq'] = 3
df_habitudes_indiv.loc[df_habitudes_indiv['collation_freq'] == 12, 'collation_freq'] = 4
df_habitudes_indiv.loc[df_habitudes_indiv['collation_freq'] == 11, 'collation_freq'] = 5
df_habitudes_indiv.loc[df_habitudes_indiv['collation_freq'] == 10, 'collation_freq'] = 0

# traitement conso bio
formatage_bio('conso_lait_bio_freq')
formatage_bio('conso_oeufs_bio_freq')
formatage_bio('conso_volailles_bio_freq')
formatage_bio('conso_viandes_bio_freq')
formatage_bio('conso_poissons_bio_freq')
formatage_bio('conso_fruits_bio_freq')
formatage_bio('conso_pdt_bio_freq')
formatage_bio('conso_legumes_bio_freq')
formatage_bio('conso_legumes_secs_bio_freq')
formatage_bio('conso_pain_bio_freq')
formatage_bio('conso_cereales_bio_freq')

# traitement variables viandes_poissons_oeufs_bio
df_habitudes_indiv['viandes_poissons_oeufs_bio'] = ((df_habitudes_indiv['conso_volailles_bio_freq'] + df_habitudes_indiv['conso_viandes_bio_freq']
+ df_habitudes_indiv['conso_oeufs_bio_freq'])/3).astype(int)

# traitement variables fruits_legumes_bio_freq
df_habitudes_indiv['fruits_legumes_bio_freq'] = ((df_habitudes_indiv['conso_fruits_bio_freq'] + df_habitudes_indiv['conso_pdt_bio_freq']
+ df_habitudes_indiv['conso_legumes_bio_freq'] + df_habitudes_indiv['conso_legumes_secs_bio_freq'])/4).astype(int)

# traitement variables pain_cereales_bio_freq
df_habitudes_indiv['pain_cereales_bio_freq'] = ((df_habitudes_indiv['conso_pain_bio_freq'] + df_habitudes_indiv['conso_cereales_bio_freq'])/2).astype(int)

# traitement valeurs manquantes - consommation_bio
df_habitudes_indiv.loc[df_habitudes_indiv['consommation_bio'].isna(), 'consommation_bio'] = 0

# variables sélectionnées
habitudes_indiv_columns = ['NOIND', 'restaurationrapide_freq', 'collation_freq', 'pain_cereales_bio_freq', 'fruits_legumes_bio_freq', 
                           'conso_lait_bio_freq', 'viandes_poissons_oeufs_bio', 'consommation_bio']

df_habitudes_indiv_select = df_habitudes_indiv[habitudes_indiv_columns] 

df_habitudes_indiv_select.rename(columns={'pain_cereales_bio_freq':'pain_cereales_bio', 'fruits_legumes_bio_freq':'fruits_legumes_bio',
                                          'conso_lait_bio_freq':'produits_laitiers_bio', 'viandes_poissons_oeufs_bio':'viandes_poissons_bio'}, inplace=True)

print('df_habitudes_indiv_select.shape :', df_habitudes_indiv_select.shape)
# print("Missing values = ", df_habitudes_indiv_select.isna().sum())
print("Total missing values = ", df_habitudes_indiv_select.isna().sum().sum())
df_habitudes_indiv_select.head()
```

## Table ACTPHYS_SEDENT: Sélection variables


```python
# Table ACTPHYS_SEDENT
path_actphys_sedent = "./datas/actphys-sedent.csv"
df_actphys_sedent = pd.read_csv(path_actphys_sedent, sep=';', error_bad_lines=False, index_col=False)

# Sélection variables
selected_columns = ['NOIND', 'tv_duree', 'ordi_duree', 'travail_duree', 'sedentarite_duree',  
                    'activite_total_duree', 'activite_domloissport_duree']
df_actphys_sedent_select = df_actphys_sedent[selected_columns]

# traitement valeurs manquantes
df_actphys_sedent_select['tv_duree'].fillna(df_actphys_sedent_select['tv_duree'].mean(), inplace=True)
df_actphys_sedent_select['ordi_duree'].fillna(df_actphys_sedent_select['ordi_duree'].mean(), inplace=True)
df_actphys_sedent_select['travail_duree'].fillna(df_actphys_sedent_select['travail_duree'].mean(), inplace=True)
df_actphys_sedent_select['sedentarite_duree'].fillna(df_actphys_sedent_select['sedentarite_duree'].mean(), inplace=True)

print('df_actphys_sedent_select.shape :', df_actphys_sedent_select.shape)
df_actphys_sedent_select.describe()
# df_actphys_sedent_select.tail()
```

## Join tables

```python
# df_description_indiv_select
# df_habitudes_indiv_select
# df_actphys_sedent_select

df_base = df_description_indiv_select.join(df_habitudes_indiv_select.set_index('NOIND'), on='NOIND')
df_base = df_base.join(df_actphys_sedent_select.set_index('NOIND'), on='NOIND')

# drop pop2 pop3 columns
df_base.drop(['pop2', 'pop3'], axis=1, inplace=True)

print('df_base.shape :', df_base.shape)
print("Total missing values = ", df_base.isna().sum().sum())
df_base[['imc','restaurationrapide_freq','collation_freq','pain_cereales_bio','fruits_legumes_bio',
              'produits_laitiers_bio','viandes_poissons_bio','consommation_bio','tv_duree','ordi_duree',
              'travail_duree','sedentarite_duree','activite_total_duree','activite_domloissport_duree']].describe()
# df_base.head()
```
